```python
string = 'AAA/BBB/CCC'
answer  = []
tmp_str = ''

for i, c in enumerate(string) :
  if c != '/' and i != len(string) - 1 :
    tmp_str += c
  elif c == '/':
    answer.append(tmp_str)
    tmp_str = ''
  elif i == len(string) - 1 :
    tmp_str += c
    answer.append(tmp_str)
answer
```

```python
def my_divide(num1, num2) :
  if not num2 :    # num2 == 0
    print('0으로 나눌 수 없습니다')
    return
  ans = num1 / num2
  print('정상적으로 나누어집니다')
  return ans

print(my_divide(10, 5))
print(my_divide(10, 0))
```

```python
def print_reverse1(string) :
  for i in range(len(string)) :
    print(string[len(string) - i - 1], end ='')
  print()

def print_reverse2(string) :
  print(string[::-1])

print_reverse1('Python1')
print_reverse2('Python2')
```

```python
def print_even(lst) :
  for num in lst :
    if not num % 2 :     # num % 2 == 0
      print(num, end = '  ')

aa = [1, 3, 2, 4, 5, 10, 17, 16]
print_even(aa)
```

```python
def print_keys(dict) :
  for key in dict.keys() :
    print(key, end = '   ')

aa = {'name' : 'Peter', 'age' : 20, 'gender' : 'M'}
print_keys(aa)
```

```python
def make_list1(string) :
  return list(string)

def make_list2(string) :
  ans = []
  for c in string :
    ans.append(c)

  return ans

print(make_list1("Python"))
print(make_list2("Artificial"))
```

```python
def max_args1(*args) :
  max_num = args[0]

  for num in args :
    if num > max_num :
      max_num = num

  return max_num

print(max_args1(-1, -2, -5, -20, -6))
print(max_args1(-1, 2, 5, 20, 6))
```

```python
def max_args2(*args) :
  return max(args)

print(max_args2(-1, -2, -5, -20, -6))
print(max_args2(-1, 2, 5, 20, 6))
```

```python
def add_mul(num1, num2) :
  return num1 + num2, num1 * num2

digit1 = int(input('첫번째 수 : '))
digit2 = int(input('두번째 수 : '))

ans1, ans2 = add_mul(digit1, digit2)
print(f'덧셈 결과는 {ans1}, 곱셈 결과는 {ans2}')
```

```python
def my_factorial(num) :
  ans = 1
  for n in range(1, num + 1) :
    ans = ans * n    # ans *= n

  return ans

my_factorial(6)
```

```python
def make_dict(string) :
  dict = {}
  slst = string.split()
  for s in slst :    # s : 'I' , 'love', 'you'
    dict[s] = len(s)

  return dict

make_dict("I eat an apple")
```

```python
def make_list(*args) :
  ans = []
  for word in args :
    ans.append(len(word))

  return ans

make_list('craze', 'little', 'things', 'call', 'love')
```

```python
def upperwords1(string) :
  words = string.split()
  answer = ''

  for word in words :
    temp = word[0].upper() + word[1:]
    answer = answer + ' ' + temp
    #print(answer)

  return answer.strip()

upperwords1('abc Def gAB')
```

```python
def upperwords2(string) :
  words = string.split()
  lst = []

  for word in words :
    temp = word[0].upper() + word[1:]
    lst.append(temp)

  return ' '.join(lst)

upperwords2('abc Def gAB')
```
