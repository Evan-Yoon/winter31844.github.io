```python
class Car :
  def init(self) :
    self.color = 'black'
    self.speed = 100

  def info(self)  :
    print(f'색깔은 {self.color}, 속도는 {self.speed}')
```

```python
redcar = Car()
bluecar = Car()

redcar.init()
redcar.info()
redcar.color = 'red'
redcar.speed = 150
redcar.info()
bluecar.color = 'blue'
bluecar.speed = 80
bluecar.info()
```

```python
class FourCal :
  def __init__(self, fir, sec) :
    self.first = fir
    self.second = sec

  def add(self) :
    return self.first + self.second

  def sub(self) :
    return self.first - self.second

  def mul(self) :
    return self.first * self.second

  def div(self) :
    return self.first / self.second
```

```python
myfourcal1 = FourCal(8, 2)
print('Add :', myfourcal1.add())
print('MUL :', myfourcal1.mul())
```

```python
class MoreFourCal(FourCal) :
  def pow(self) :
    return self.first ** self.second

mymore1 = MoreFourCal(4, 3)
print('Add :', mymore1.add())
print('MUL :', mymore1.mul())
print('POW :', mymore1.pow())
```

```python
class SafeFourCal(FourCal) :
  def div(self) :
    if self.second == 0 :
      return 0
    else :
      return self.first / self.second

mysafefourcal1 = SafeFourCal(4, 2)
print(mysafefourcal1.div())

mysafefourcal2 = SafeFourCal(4, 0)
print(mysafefourcal2.div())

myfourcal3 = FourCal(4, 0)
print(myfourcal3.div())
```

```python
def new_div(a, b) :
  try :
    result = a / b
    print('Result :', result)
  except ZeroDivisionError :
    print('Error : Division by zero')
  except TypeError :
    print('Error : Type Error')
  finally :
    print('실행종료')
```

```python
new_div(4, 2)
new_div(3, 0)
new_div(4, '가')
```

```python
import random

class Account :
  number_of_account = 0

  def __init__(self, name, balance) :
    Account.number_of_account += 1
    self.bank_name = 'KB Bank'
    self.name = name
    self.balance = balance

    num1 = random.randint(0, 999)
    num2 = random.randint(0, 99)
    num3 = random.randint(0, 999999)

    num1 = str(num1).zfill(3)
    num2 = str(num2).zfill(2)
    num3 = str(num3).zfill(6)

    self.account_num = num1 + '-' + num2 + '-' + num3

  def deposit(self, money) :
    if money >= 1 :
      self.balance += money
      print(f'현재 잔액은 {self.balance}원 입니다')

    else :
      print('입금액 오류 입니다.')

  def withdraw(self, money) :
    if self.balance >= money :
      self.balance -= money
      print(f'현재 잔액은 {self.balance}원 입니다')
    else :
      print('잔액이 부족합니다.')

  def display_info(self) :
    print('은행명 :', self.bank_name)
    print('고객명 :', self.name)
    print('계좌번호 :', self.account_num)
    print('잔액 :', self.balance)
```

```python
my_acc1 = Account('홍길동', 10000)
my_acc2 = Account('김철수', 5000)

my_acc1.deposit(0)       # 입금오류
my_acc1.deposit(1000)  # 잔액 11000
my_acc2.withdraw(6000)   # 잔액 부족
my_acc2.withdraw(1000)   # 잔액 4000

my_acc1.display_info()
my_acc2.display_info()

print(f'현재까지 생성된 계좌의 수는 {Account.number_of_account}개 입니다')
```

```python
x = np.linspace(0, 2 * np.pi, 400)
y1 = np.sin(x**2)
y2 = np.cos(x**2)

fig, axes = plt.subplots(2, 2)
axes[0,0].plot(x, y1, 'tab:blue')
axes[0,0].set_title('Axes[0,0]')
axes[0,1].plot(x, y2, 'tab:orange')
axes[0,1].set_title('Axes[0,1]')
axes[1,0].plot(x, -y1, 'tab:green')
axes[1,0].set_title('Axes[1,0]')
axes[1,1].plot(x, -y2, 'tab:red')
axes[1,1].set_title('Axes[1,1]')

for ax in axes.flat :
  ax.set(xlabel='x-axis', ylabel='y-axis')
  ax.label_outer()

plt.show()
```
